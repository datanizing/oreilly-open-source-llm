# Text generation with TGI from Hugging Face

[TGI](https://huggingface.co/docs/text-generation-inference/index) was one of the
first text generation servers and created by Hugging Face in Python and Rust.

However, even if TGI is still available, Hugging Face recommends using `vllm` 
which does not need `docker` etc. As both servers have an OpenAI compatible
API, swapping the backends is very easy.
